```python
import numpy as np
import os
import matplotlib.pyplot as plt
import random
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from dipy.io.image import load_nifti_data
from typing import Union
from tqdm import tqdm
```

```python
class RATLoaderExplorer:
    '''
    A partir del archivo nii vamos a hacer el análisis de los datos de la rata
    para generar los mapas de T1,T2 y M0
    '''
    def __init__(self,file : str,mask: str):
        '''
        Se necesita una máscara ya que solo queremos procesar lo que concierne a cerebros,
        lo demás como cráneo y cosas del estilo no es necesario
        '''
        self.data=load_nifti_data(file)
        self.mascara=load_nifti_data(mask)
        self.mascara=self.mascara[:,:,0]
        self.__TRep=np.array([5000,3000,1500,800,400,200])
        self.__TEch=np.linspace(7,63,num=5,dtype=int)
        
    def return_slice(self,slindex: int):
        try: 
            return self.data[:,:,slindex]
        except:
            raise ValueError("Indice de la rebanada fuera de rango")
    
    def return_voxel_signal(self,index: tuple,reptime: Union[int,tuple,None] =None, echos: Union[int,tuple,None]=None):
        '''
        En este caso necesitamos una pareja de números x,y que se mandan como argumento
        '''
        if reptime is None and echos is None:
            return self.data[index[0],index[1],:]
        
        elif type(reptime) == int and echos is None:
            '''
            Regresamos todas las observaciones de un determinado tiempo de repetición
            '''
            ind=np.where(self.__TRep == reptime)
            return self.data[index[0],index[1],:ind[0][0]+5]
        elif type(reptime) ==tuple and echos is None:
            ind1=np.isin(self.__TRep,np.asarray(reptime))
            ind=np.where(ind1==True)[0]*5
            print(ind)
            signal=[]
            for j in range(ind.shape[0]):
                signal=signal+self.data[index[0],index[1],ind[j]:ind[j]+5].tolist()
            return np.array(signal)
        
        elif reptime is None and type(echos)==int:
            ind=np.where(self.__TEch==echos)[0]
            signal=[]
            for i in range(6):
                signal.append(self.data[index[0],index[1],5*i+ind[0]])
            signal.reverse()
            return np.array(signal)
        
        elif type(reptime)==int and type(echos)==int:
            indr=np.where(self.__TRep==reptime)[0][0]
            ind=np.where(self.__TEch==echos)[0][0]
            return self.data[index[0],index[1],5*indr+ind]
        else:
            '''
            No se ocupan mas casos
            '''
            pass
    
    def T2_by_boxel(self,index: tuple):
        '''
        Recordemos que para obtener el mayor contraste de impagenes T2 necesitamos
        TR largo entonces en este caso nos tomaremos toda la señal con TR=5 segundos
        '''
        y_data=self.return_voxel_signal(index,reptime=5000)
        x_data=self.__TEch
        try:
            popt,pcov=curve_fit(self.T2,x_data,y_data,bounds=(0,200))
        except:
            popt=(1.0,1.0)
        M0,T2=popt
        return M0,T2 
    
    def T1_by_boxel(self,index: tuple):
        '''
        En el caso del cálculo de T1 necesitaremos todos los tiempos de repetición y el tiempo echo mas corto
        pero en orden inverso
        '''
        y_data=self.return_voxel_signal(index,echos=7)
        x_data=self.__TRep
        x_data=np.flip(x_data,axis=0)
        try:
            popt,pcov=curve_fit(self.T1,x_data,y_data,bounds=(0, 4500))
        except:
            popt=(0,0) #No se encontraron datoa suficientes para hacer el ajuste de curvas
        M0,T1=popt
        return M0,T1
    
    def PD_by_voxel(self,index: tuple):
        '''
        Recordemos que para el pesado por densidad de protones de hidrógeno 
        '''
        return self.return_voxel_signal(index,reptime=5000,echos=7)
    
    def ConstructMap(self,type: str):
        '''
        Construcción de los diferentes mapas de la tarea,
        tenemos por densidad de protones (pd), valores t1 (t1), t2(t2) y
        pesado por intensidad ("Le llamamos así a la imágen con el tiempo de repetición más largo y el tiempo eco más corto")
        '''
        if type not in ['pd','t1','t2','inte']:
            raise ValueError("Tipo de mapa no soportado")
        mask=np.zeros_like(self.data[:,:,0]) #Solo la primera rebanada
        if type=='inte':
            for i in tqdm(range(mask.shape[0])):
                for j in range(mask.shape[1]):
                    if self.mascara[i][j]!=0:
                        mask[i][j]=self.PD_by_voxel((i,j))
        if type=='pd':
            for i in tqdm(range(mask.shape[0])):
                for j in range(mask.shape[1]):
                    if self.mascara[i][j]!=0:
                        mask[i][j],_=self.T2_by_boxel((i,j))
        if type=='t2':
            for i in tqdm(range(mask.shape[0])):
                for j in range(mask.shape[1]):
                    if self.mascara[i][j]!=0:
                        _,mask[i][j]=self.T2_by_boxel((i,j))
        if type=='t1':
            for i in tqdm(range(mask.shape[0])):
                for j in range(mask.shape[1]):
                    if self.mascara[i][j]!=0:
                        _,mask[i][j]=self.T1_by_boxel((i,j))
        return mask
    
    def stats_(self,tipo:'str',return_points:bool=False):
        valores=self.ConstructMap(type=tipo)
        new_vals=valores[np.nonzero(valores)]
        if not return_points:
            return np.mean(new_vals),np.std(new_vals)
        else:
            return np.mean(new_vals),np.std(new_vals),new_vals

    @staticmethod
    def T1(t,a,b):
        '''
        Fórmula para el T1
        '''
        return a*(1-np.exp(-t/b))
    @staticmethod
    def T2(t,a,b):
        '''
        Fórmula para el T2
        ''' 
        return a*np.exp(-t/b)
```

```python
rle=RATLoaderExplorer('./20181018_140750T1T2mapRAREs60001a001.nii','./Mask_segmentation.nii')
```

# Calcular los mapas

## Mapa Pesado en Intensidad

```python
pd=rle.ConstructMap(type='inte')
```

## Mapa pesado en T1

```python
pdt1=rle.ConstructMap(type='t1')
```

# Mapa pesado en T2

```python
pdt2=rle.ConstructMap(type='t2')
```

# Mapa pesado en Densidad de Protón

```python
pdm0=rle.ConstructMap(type='pd')
```
