```python
import numpy as np
import os 
import re
import matplotlib.pyplot as plt
```

```python
f=lambda x: x[0]*np.exp(-x[1]/x[2])

g=lambda x: x[0]*(1-np.exp(-x[1]/x[2]))

def build_values_t2(M0,T2,trange=np.linspace(10,2000,num=50)):
    return trange,[f((M0,t,T2)) for t in trange]

def build_values_t1(M0,T1,trange=np.linspace(10,2000,num=50)):
    return trange,[g((M0,t,T1)) for t in trange]
```

```python
rango,valores=build_values_t2(100,2000)
rango1,valores1=build_values_t2(100,100)
rango2,valores2=build_values_t2(100,1000)
rango3,valores3=build_values_t2(100,300)
plt.plot(rango,valores,label='botella 3')
plt.plot(rango1,valores1,color='red',label='botella 1')
plt.plot(rango2,valores2,color='magenta',label='botella 4')
plt.plot(rango3,valores3,color='green',label='botella 2')
plt.plot([90 for i in range(len(rango1))],np.linspace(0,100,num=50))
plt.legend()
plt.title("Caidas T2 de las diferentes botellas con M0=100")
plt.xlabel("Tiempo en ms")
plt.savefig("Intensidades_T2_te.jpg")
```

```python
rango,valores=build_values_t1(100,4000)
rango1,valores1=build_values_t1(100,280)
rango2,valores2=build_values_t1(100,2500)
rango3,valores3=build_values_t1(100,600)
plt.plot(rango,valores,label='botella 3')
plt.plot(rango1,valores1,color='red',label='botella 1')
plt.plot(rango2,valores2,color='magenta',label='botella 4')
plt.plot(rango3,valores3,color='green',label='botella 2')
plt.plot([500 for i in range(len(rango1))],np.linspace(0,100,num=50))
plt.legend()
plt.title("Caidas T1 de las diferentes botellas con M0=100")
plt.xlabel("Tiempo en ms")
plt.savefig("Intensidades_T1_tr.jpg")
```
